### There is no need of Fixing Wrong Data or Handling outliers before performing factor analysis as all values ranges from 3 to 10

```python
# Perform Bartlett’s test of sphericity that checks whether or not the observed variables intercorrelate
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value, p_value= calculate_bartlett_sphericity(df)
chi_square_value, p_value
```

### The small p-value indicates that the observed correlations are unlikely to have occurred by chance alone.
### This result suggests that factor analysis could be appropriate for our dataset, as it assumes some level of correlation among variables to identify underlying factors.

```python
# Perform Kaiser-meyer-olkin (kmo) test that measures the suitability of data for factor analysis.
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all, kmo_model= calculate_kmo(df)
kmo_all, kmo_model
```

### It quantifies the proportion of variance among variables that might be common variance. A KMO value closer to 1 suggests that the data is more suitable for factor analysis, while lower values indicate that the data may not be suitable.

### Overall KMO value (kmo_model): This is the overall measure of sampling adequacy for the factor analysis, In our case, the overall KMO value of approximately 0.798 suggests that the dataset has a reasonably high degree of suitability for factor analysis.

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer(12, rotation=None)

# fit the model to the data
fa.fit(df)
```

### The 12 parameter specifies the number of factors to extract, and rotation=none specifies that no rotation should be performed.

```python
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev
```

### Factors with eigenvalues greater than 1 are considered significant and should be retained, while factors with eigenvalues less than 1 can be dropped.

```python
# Create a scree plot using matplotlib
plt.scatter(range(1, df.shape[1]+1), ev)
plt.plot(range(1,df.shape[1]+1), ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

### creating a scree plot with markers and a connected line using plt.scatter and plt.plot to visually inspect the eigenvalues and determine the appropriate number of factors to retain.
### Based On that I see that we should choose to retain the first 4 factors and consider dropping the remaining factors
